## **Princípios gerais para o uso de cores em plotagens**
### **Componentes de cor**
Devido à forma como nossos olhos funcionam, uma cor particular pode ser definida usando três componentes. Normalmente programamos as cores em um computador especificando seus valores RGB, que definem a intensidade dos canais <font color = 'red'>**vermelho**</font>, <font color = 'green'>**verde**</font> e <font color = 'blue'>**azul**</font> em uma tela. Mas para analisar os atributos perceptivos de uma cor, é melhor pensar em termos de canais de matiz, saturação e luminância.

Matiz é o componente que distingue “cores diferentes” em um sentido não técnico. É a propriedade da cor que leva a nomes de primeira ordem como "vermelho" e "azul":
<img src="img/HUE.png">

Saturação (ou croma) é o colorido. Duas cores com matizes diferentes parecerão mais distintas quando tiverem mais saturação:
<img src="img/Saturation.png">

E a luminosidade corresponde a quanta luz é emitida (ou refletida, para cores impressas), variando do preto ao branco:
<img src="img/Lightness.png">

### **Varie o matiz para distinguir as categorias**
Quando você deseja representar várias categorias em um gráfico, normalmente deve variar a cor dos elementos. Considere este exemplo simples: em qual desses dois gráficos é mais fácil contar o número de pontos triangulares?

<img src="img/Triangular points.png">

No gráfico da direita, os triângulos laranja “saltam”, tornando mais fácil distingui-los dos círculos. Esse efeito pop-out ocorre porque nosso sistema visual prioriza as diferenças de cores.

As cores azul e laranja diferem principalmente em termos de matiz. O matiz é útil para representar categorias: a maioria das pessoas consegue distinguir um número moderado de matizes com relativa facilidade, e os pontos com matizes diferentes, mas com brilho ou intensidade semelhantes, parecem igualmente importantes. Também torna mais fácil falar sobre os enredos. Considere este exemplo:

<img src="img/Colors Category.png">

A maioria das pessoas seria capaz de determinar rapidamente que existem cinco categorias distintas no gráfico à esquerda e, se solicitadas a caracterizar os pontos “azuis”, seria capaz de fazê-lo.

Com o gráfico à direita, onde os pontos são todos azuis, mas variam em luminância e saturação, é mais difícil dizer quantas categorias exclusivas estão presentes. E como falaríamos sobre uma determinada categoria? "Os pontos razoavelmente-mas-não-muito-azuis?" Além do mais, os pontos cinza parecem desaparecer no fundo, sem ênfase em relação aos pontos azuis mais intensos. Se as categorias forem igualmente importantes, esta é uma representação pobre.

Portanto, como regra geral, use a variação de matiz para representar as categorias. Com isso dito, aqui estão algumas notas de cautela. Se você tiver mais do que um punhado de cores em sua trama, pode ser difícil ter em mente o que cada uma significa, a menos que haja associações pré-existentes entre as categorias e as cores usadas para representá-las. Isso torna seu gráfico mais difícil de interpretar: em vez de focar nos dados, o visualizador terá que consultar continuamente a legenda para entender o que é mostrado. Portanto, você deve se esforçar para não fazer enredos muito complexos. E esteja ciente de que nem todo mundo vê as cores da mesma maneira. A variação da forma (ou algum outro atributo) e cor pode ajudar as pessoas com visão de cores anômala a entender seus gráficos e pode mantê-los (de alguma forma) interpretáveis ​​se forem impressos em preto e branco.

### **Varie a luminância para representar números**
Por outro lado, as variações de matiz não são adequadas para representar dados numéricos. Considere este exemplo, onde precisamos de cores para representar as contagens em um histograma bivariado. À esquerda, usamos um mapa de cores circular, onde mudanças graduais no número de observações dentro de cada caixa correspondem a mudanças graduais no matiz. À direita, usamos uma paleta que usa cores mais brilhantes para representar caixas com contagens maiores:

<img src="img/Class Ramp Color.png">

Com a paleta baseada em matiz, é muito difícil determinar a forma da distribuição bivariada. Em contraste, a paleta de luminância torna muito mais claro que existem dois picos proeminentes.

A variação da luminância ajuda a ver a estrutura dos dados, e as mudanças na luminância são processadas de forma mais intuitiva como mudanças na importância. Mas o gráfico à direita não usa um mapa de cores em tons de cinza. Seu colorido o torna mais interessante, e a variação sutil de matiz aumenta a distância perceptiva entre dois valores. Como resultado, pequenas diferenças são ligeiramente mais fáceis de resolver.

Esses exemplos mostram que as opções de paleta de cores são mais do que estéticas: as cores que você escolhe podem revelar padrões em seus dados, se usados ​​de forma eficaz, ou ocultá-los, se usados ​​de maneira inadequada. Não existe uma paleta ideal, mas existem paletas que são melhores ou piores para conjuntos de dados específicos e abordagens de visualização.

E a estética é importante: quanto mais as pessoas quiserem ver suas figuras, maior será a chance de aprenderem algo com elas. Isso é verdade mesmo quando você está traçando para si mesmo. Durante a análise exploratória de dados, você pode gerar muitos números semelhantes. Variar as paletas de cores adicionará uma sensação de novidade, o que o mantém envolvido e preparado para observar recursos interessantes de seus dados.

Então, como você pode escolher paletas de cores que representem bem seus dados e tenham uma aparência atraente?

## **Princípios gerais para o uso de cores em plotagens**
### **Componentes de cor**

A função mais importante para trabalhar com paletas de cores é, apropriadamente, color_palette (). Esta função fornece uma interface para a maioria das maneiras possíveis de gerar paletas de cores em países do mar. E é usado internamente por qualquer função que tenha um argumento de paleta.

O principal argumento para [color_palette ()](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette) é geralmente uma string: ou o nome de uma paleta específica ou o nome de uma família e argumentos adicionais para selecionar um membro específico. No último caso, [color_palette ()](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette) delegará a uma função mais específica, como [cubehelix_palette ()](https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette). Também é possível passar uma lista de cores especificadas de qualquer maneira aceita por matplotlib (uma tupla RGB, um código hexadecimal ou um nome na tabela X11). O valor de retorno é um objeto que envolve uma lista de tuplas RGB com alguns métodos úteis, como conversão para códigos hexadecimais e uma representação HTML rica.

Chamar [color_palette ()](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette) sem argumentos retornará a paleta de cores padrão atual que matplotlib (e a maioria das funções marítimas) usará se as cores não forem especificadas de outra forma. Esta paleta padrão pode ser definida com a função [set_palette ()](https://seaborn.pydata.org/generated/seaborn.set_palette.html#seaborn.set_palette) correspondente, que chama [color_palette ()](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette) internamente e aceita os mesmos argumentos.

Para motivar as diferentes opções que color_palette () oferece, será útil apresentar um esquema de classificação para paletas de cores. De modo geral, as paletas se enquadram em uma das três categorias:

* paletas qualitativas, boas para representar dados categóricos

* paletas sequenciais, boas para representar dados numéricos

* paletas divergentes, boas para representar dados numéricos com um limite categórico



## **Paletas de cores qualitativas**
Paletas qualitativas são adequadas para representar dados categóricos porque a maior parte de sua variação está no componente matiz. A paleta de cores padrão no seaborn é uma paleta qualitativa com dez tons distintos:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline # Graficos gerados na linha
```

```python
# base de amostra
x = np.arange(10000).reshape(100,100)
```

```python
sns.color_palette()
```

Essas cores têm a mesma ordem da paleta de cores matplotlib padrão, "tab10", mas são um pouco menos intensas. Comparar:

```python
sns.color_palette("tab10")
```

Seaborn, na verdade, tem seis variações da paleta de matplotlib, chamadas de deep, muted, pastel, bright, dark e color daltônico. Eles abrangem uma faixa de valores médios de luminância e saturação:

<img src="Img/Matplotlib’s Palette.png">

Muitas pessoas consideram os matizes moderados da paleta "profunda" padrão esteticamente agradáveis, mas também são menos distintos. Como resultado, eles podem ser mais difíceis de discriminar em alguns contextos, o que é algo a se ter em mente ao fazer gráficos de publicação. Essa comparação pode ser útil para estimar o desempenho das paletas de cores marinhas ao simular diferentes formas de daltonismo.

### **Usando sistemas de cores circulares**

Quando você tem um número arbitrário de categorias, a abordagem mais fácil para encontrar matizes exclusivos é desenhar cores com espaçamento uniforme em um espaço de cores circular (onde o matiz muda enquanto mantém o brilho e a saturação constantes). É para isso que a maioria das funções marítimas usa o padrão quando precisam usar mais cores do que as definidas atualmente no ciclo de cores padrão.

A maneira mais comum de fazer isso usa o espaço de cores hls, que é uma transformação simples dos valores RGB. Vimos essa paleta de cores antes como um contra-exemplo de como plotar um histograma:

```python
sns.color_palette("hls", 8)
```

Devido à forma como o sistema visual humano funciona, as cores que têm a mesma luminância e saturação em termos de seus valores RGB não parecerão necessariamente igualmente intensas. que atinge menos variação de intensidade conforme você gira em torno da roda de cores:

```python
sns.color_palette("husl", 8)
```

Quando o mar precisa de uma paleta categórica com mais cores do que as disponíveis no padrão atual, ele usará esta abordagem.

### **Usando paletas categóricas do Color Brewer**

Outra fonte de paletas categóricas visualmente agradáveis vem da ferramenta Color Brewer (que também possui paletas sequenciais e divergentes, como veremos abaixo).

```python
sns.color_palette("Set2")
```

Esteja ciente de que as paletas qualitativas do [Color Brewer](https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3) têm comprimentos diferentes, e o comportamento padrão de [color_palette ()](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette) é fornecer a lista completa:

```python
sns.color_palette("Paired")
```

## **Paletas de cores sequenciais**
A segunda classe principal de paletas de cores é chamada de “sequencial”. Esse tipo de mapeamento é apropriado quando os dados variam de valores relativamente baixos ou desinteressantes a valores relativamente altos ou interessantes (ou vice-versa). Como vimos acima, a dimensão primária da variação em uma paleta sequencial é a luminância. Algumas funções marítimas serão padronizadas para uma paleta sequencial quando você estiver mapeando dados numéricos. (Por razões históricas, os mapeamentos categóricos e numéricos são especificados com o parâmetro hue em funções como [relplot ()](https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot) ou [displot ()](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot), embora os mapeamentos numéricos usem paletas de cores com relativamente pouca variação de matiz).

### **Paletas perceptivamente uniformes**
Como se destinam a representar valores numéricos, as melhores paletas sequenciais serão _perceptualmente uniformes_, o que significa que a discriminabilidade relativa de duas cores é proporcional à diferença entre os valores de dados correspondentes. Seaborn inclui quatro mapas de cores sequenciais perceptualmente uniformes: "foguete", "mako", "flare" e "crista". Os dois primeiros têm uma faixa de luminância muito ampla e são adequados para aplicações como mapas de calor, onde as cores preenchem o espaço em que são plotadas:

```python
cmap = sns.color_palette("rocket", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

```python
cmap = sns.color_palette("mako", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Como os valores extremos desses mapas de cores se aproximam do branco, eles não são adequados para colorir elementos como linhas ou pontos: será difícil discriminar valores importantes contra um fundo branco ou cinza. Os mapas de cores “flare” e “crest” são a melhor escolha para tais plotagens. Eles têm uma gama mais restrita de variações de luminância, que compensam com uma variação um pouco mais pronunciada de matiz. A direção padrão da rampa de luminância também é invertida, para que valores menores tenham cores mais claras:

```python
cmap = sns.color_palette("flare", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

```python
cmap = sns.color_palette("crest", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Também é possível usar os mapas de cores perceptualmente uniformes fornecidos pelo matplotlib, como "magma" e "viridis"

```python
cmap = sns.color_palette("magma", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

```python
cmap = sns.color_palette("viridis", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Tal como acontece com a convenção em matplotlib, todo mapa de cores contínuo tem uma versão invertida, que tem o sufixo "_r":

```python
cmap = sns.color_palette("rocket_r", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

### **Mapeamento discreto vs. contínuo**
Uma coisa a estar ciente é que os oceanos podem gerar valores discretos a partir de mapas de cores sequenciais e, ao fazer isso, não usará os valores mais extremos. Compare a versão discreta de "foguete" com a versão contínua mostrada acima:


```python
sns.color_palette("rocket")
```

Internamente, o seaborn usa a versão discreta para dados categóricos e a versão contínua quando no modo de mapeamento numérico. Mapas de cores sequenciais discretos podem ser adequados para visualizar dados categóricos com uma ordem intrínseca, especialmente se houver alguma variação de matiz.

### **Paletas sequenciais “cubehelix ”**
Os mapas de cores perceptualmente uniformes são difíceis de gerar programaticamente, porque não se baseiam no espaço de cores RGB. O sistema cubehelix oferece um compromisso baseado em RGB: ele gera paletas sequenciais com um aumento ou diminuição linear no brilho e alguma variação contínua no matiz. Embora não sejam perfeitamente uniformes em termos de percepção, os mapas de cores resultantes têm muitas propriedades boas. É importante ressaltar que muitos aspectos do processo de design são parametrizáveis.

Matplotlib tem a versão padrão do cubehelix incorporada a ele:

```python
cmap = sns.color_palette("cubehelix", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

A paleta padrão retornada pela função seaborn [cubehelix_palette ()](https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette) é um pouco diferente do padrão matplotlib por não girar tanto em torno da roda de matiz ou cobrir uma ampla gama de intensidades. Ele também inverte a rampa de luminância:

```python
cmap = sns.cubehelix_palette(as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Outros argumentos para [cubehelix_palette ()](https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette) controlam a aparência da paleta. As duas coisas principais que você mudará são o início (um valor entre 0 e 3) e a rotação, ou número de rotações (um valor arbitrário, mas geralmente entre -1 e 1)

```python
cmap = sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Quanto mais você gira, mais variação de matiz você verá:

```python
cmap = sns.cubehelix_palette(start=.5, rot=-.75, as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Você pode controlar o quão escuro e claro os pontos de extremidade são e sua ordem:

```python
cmap = sns.cubehelix_palette(start=2, rot=0, dark=0, light=.95, reverse=True, as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

### **Paletas sequenciais personalizadas**
Para obter uma interface mais simples para paletas sequenciais personalizadas, você pode usar [light_palette ()](https://seaborn.pydata.org/generated/seaborn.light_palette.html#seaborn.light_palette) ou [dark_palette ()](https://seaborn.pydata.org/generated/seaborn.dark_palette.html#seaborn.dark_palette), que são propagadas com uma única cor e produzem uma paleta que rampa de valores desaturados claros ou escuros para essa cor:


```python
cmap = sns.light_palette("seagreen", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

```python
cmap = sns.dark_palette("#69d", reverse=True, as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Assim como acontece com as paletas cubehelix, você também pode especificar paletas claras ou escuras por meio de [color_palette ()](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette) ou em qualquer lugar que a paleta seja aceita:

```python
cmap = sns.color_palette("light:b", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

Inverta o mapa de cores adicionando "_r":

```python
cmap = sns.color_palette("dark:salmon_r", as_cmap=True)
ax = sns.heatmap(x, cmap=cmap)
```

The Color Brewer library also has some good options for sequential palettes. They include palettes with one primary hue:

Junto com opções de vários matizes:

```python
cmap = sns.color_palette("YlOrBr", as_cmap=True)
x = sns.heatmap(x, cmap=cmap)
```

## **Paletas de cores divergentes**
A terceira classe de paletas de cores é chamada de “divergentes”. Eles são usados para dados em que os valores altos e baixos grandes são interessantes e abrangem um valor de ponto médio (geralmente 0) que deve ser subdimensionado. As regras para escolher boas paletas divergentes são semelhantes às boas paletas sequenciais, exceto que agora deve haver dois matizes dominantes no mapa de cores, um em (ou próximo a) cada pólo. Também é importante que os valores iniciais sejam de brilho e saturação semelhantes.

### **Paletas divergentes perceptivamente uniformes**
Seaborn includes two perceptually uniform diverging palettes: "vlag" and "icefire". They both use blue and red at their poles, which many intuitively processes as “cold” and “hot”:

```python
cmap = sns.color_palette("vlag", as_cmap=True)
```
